## 과제 Quiz Review

```python
# Quiz 1
from nltk import book
Thursday = book.text9
Monty = book.text6

diversity_txt9 = round(len(set(Thursday.tokens)) / len(Thursday.tokens) * 100)
print(f'Text 9\'s lexical diversity is approximately {diversity_txt9}%.')

text6_tokens = sorted((set(Monty.tokens)), reverse = True) [:10]
print(text6_tokens) #첫번째 조건 추출

for text6_alligned in range(10):
    if 'z' in text6_tokens[text6_alligned]: #z가 있을 때
        print(text6_tokens[text6_alligned].capitalize())
    elif len(text6_tokens) >= 4: #z가 없을 때 4글자 이상일 때
        temporary = (text6_tokens[text6_alligned]) [::-1]
        temporary = temporary.capitalize()
        temporary = temporary[::-1]
        print(temporary)
    else: #z도 없고 4자 이상도 아닐 때 출력
        print(text6_tokens[text6_alligned])
```

#### List Comprehension
- 포함형 / 지능형 / 축약형 리스트
- 하나의 리스트에서 그것을 다른 리스트를 포함 다른 리스트를 만드는 것
- list + for loop를 활용
- `[A for B in C]`
    - C에 있는 B에서 A를 가져와라 → 리스트에 담아서 변환
    - A: 내가 원하는 출력 형태
    - B: C의 형태
    - C: 반복 가능한 자료형(리스트)
- `[A for B in C if ...]` : 조건문을 추가한 버전

```python
[phone[0] for phone in phones]
```

### NLTK - Frequency Distribution
- 단어가 몇 번 출현했는지(빈도 수)
- 텍스트의 성질을 알기 위해
    - 단어 개수
    - 단어 빈도수

```python
from nltk.probability import FreqDist # 이렇게 라이브러리를 가져와야 실행됐다.
fdist = FreqDist(book.text1)
```

```python
fdist # ''안에 있는 내용이 몇번 나왔는지
```

```python
fdist.most_common(10) # 가장 많이 나온 순서대로
```

```python
fdist.freq('that')
```

```python
fdist.plot(10, cumulative=True) # 누적해서
```

```python
fdist.plot(10, cumulative=False) #각각의
```

```python
# 한번만 나타난 단어들...
    # 알파벳만 리스트에 담아보기
lists = fdist.hapaxes()
set([word for word in fdist.hapaxes() if word.isalpha()])
```
