## 중첩조건문
- nested conditional 
- if 블록 안에 또다른 if 블록이 있는 경우
- 들여쓰기로 논리를 구분한다.
- 그렇다고 해서 elif나 논리연산자를 활용할 수 있는데도 남발하지는 말자.

```python
Latest_Version = 11
User_Version = 10
if Latest_Version == User_Version:
    print("Early Adopter")
else:
    if Latest_Version > User_Version :
        print("Normal User")
    else:
        print("Error")
#안좋은 예 (if, elif, else 사용이 적절함.)
```

```python
if x > 0:
    if x < 10:
        print("Block A")
    else:
        print("Block B")
else:
    print("block C")
# 안좋은 예, and 조건을 사용하면 좋음
# 파이썬에서는 0 < x < 10이라 표현하는게 가능
```

### 논리연산자
- 비교연산자가 여러번 사용될 때 사용하는 것
- and, or, not
    - A and B: A와 B 둘 다 참일 때 True (A & B로 사용)
    - A or B: A, B 둘 중 하나만 참일 때 True (A | B로 사용)
    - not A: A가 False(True)일 때 True(False)
- 비교연산자의 괄호는 가독성을 위해 사용(Optional)
- 파이썬에서만 가능한 표현: i.e 0 < x < 12

## `in`
- membership operator (멤버 연산자)
- 어떤 요소가 그 안에 있는지 확인하고 싶을 때 사용
- 비슷한 조건이 여러번 반복될 때 사용 가능

```python
Class_letter = 'Advanced Python'

if Class_letter in ['a','e','i','o','u']: #리스트 요소에 letter과 일치하는 것이 있는가?
    print("Yes")
else:
    print("No")
```

```python
word = 'abcde'
vowel = ['a', 'e', 'i', 'o', 'u'] 

if word in vowel:
    print("Yes")
else:
    print("No")
```

### 바다코끼리 연산자
- warlus operator
- 할당과 테스트를 동시에 하는 기능
- ':='로 사용

```python
tweet_limit = 200
tweet_string = 'blah' * 50 #공백없이 스트링 200개
tweet_string
```

```python
diff = tweet_limit - len(tweet_string)

if diff > 0:
    print("불가능")
else:
    print("불가능")
```

```python
if diff := tweet_limit - len(tweet_string) > 0:
    print("가능")
else:
    print("불가능")
```

```python
#첫번째 퀴즈 내가 쓴 답
year = int(input("연도를 입력하세요: "))

if year % 4 == 0:
    if year % 400 == 0:
        print(f'{year} 년은 윤년입니다.')
    elif year % 100 == 0:
        print(f'{year} 년은 윤년이 아닙니다.')
    else:
       print(f'{year} 년은 윤년입니다')
else:
    print(f'{year} 년은 윤년이 아닙니다.')
```

## Escape Character | 이스케이프 문자
- 눈에 보이는대로 인식되지 않는 문자열
- `\` + <character>의 형태로 작성 ex) `\n`, `\t`, `\b`, `\'`, `\"`
- print 함수를 쓰지 않으면 형태가 그대로 출력된다.
- '\n'의 출력은 어떻게?
    - `\\n`
    - `r''` (raw string(원시 문자열))
- 문자열을 바꾸는 것이 아니라 표면상의 형태를 말하는 것

```python
print('Hello\nWorld')
print("\'Iphone\' 13  \bPro")
```

### Slicing
- 지정 범위의 문자열을 추출하는 것
- 문법: 대괄호 + 오프셋 이용 `[offset]`
 - offset?: `[start:end:step]`으로 구성
- offset은 항상 0부터 시작
- start는 포함하고, end는 불포함 (크거나 같고, 작음)
    - 0이 있다면 생략해도 좋다.
        - 4까지 출력하라라 → [:5] = [0:5]
        - 처음부터 끝까지 2스텝씩 → [::2]
    - 공백도 포함한다
    - 거꾸로도 가능하다. (대신 -1부터 시작한다.)

```python
sent = 'I am learning Python'
sent, sent[3], sent[::2], sent[::-1]
```
